```python
!pip install --upgrade gradio diffusers torch torchvision torchaudio huggingface_hub --quiet

import os
import torch
import gradio as gr
from PIL import Image
from huggingface_hub import login
from diffusers import StableDiffusionImg2ImgPipeline

# STEP 1: Authenticate with Hugging Face
os.environ.pop("HF_TOKEN", None)  # Clear any stored token
login()  # Prompts for authentication

# STEP 2: Load Stable Diffusion Model ONCE
device = "cuda" if torch.cuda.is_available() else "cpu"

def load_model():
    """Loads Stable Diffusion img2img model for AI-stylization."""
    model_id = "runwayml/stable-diffusion-v1-5"
    try:
        pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
            model_id, torch_dtype=torch.float16 if device == "cuda" else torch.float32
        )
        pipe.to(device)
        print(f"✅ Model Loaded on {device}")
        return pipe
    except Exception as e:
        print(f"❌ Error loading model: {e}")
        return None

pipe = load_model()  # Load model ONCE

# STEP 3: Image Processing Function
def transform_image(uploaded_image, prompt="Studio Ghibli anime style, high quality, vibrant colors", strength=0.65):
    """Transforms an uploaded image into an AI-generated version using a given prompt."""
    if pipe:
        init_image = Image.open(uploaded_image).convert("RGB")
        init_image = init_image.resize((512, 512))  # BETTER quality

        output = pipe(
            prompt=prompt, image=init_image, strength=strength, guidance_scale=7.5
        )
        return output.images[0]
    else:
        return "❌ Error: Model not loaded."

# STEP 4: Gradio Interface
gr.Interface(
    fn=transform_image,
    inputs=gr.Image(type="filepath"),
    outputs="image",
    title="AI Ghibli Art Generator",
    description="Upload an image & convert it into a high-quality Studio Ghibli-style artwork.",
).launch(share=True)  # Enables public access
```
